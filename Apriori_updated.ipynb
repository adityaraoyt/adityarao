```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```
store_data = pd.read_csv('shopping_data.csv')
```

```
store_data.head()
```

```
  store_data.shape
```

```
records = []
for i in range(0,411):
  records.append([str(store_data.values[i,j]) for j in range(0,12)])
```

```
records[0]
```

```
final_list = []
for i in range(len(records)):
  temp = []
  for x in records[i]:
    if str(x)!='nan':
      temp.append(x)
  final_list.append(temp)
```

```
final_list[0]
```

```
association_rules = apriori(final_list,min_support = 0.0045,min_confidence = 0.2, min_lift = 3,min_length = 2)
association_results = list(association_rules)
```

```
len(association_results)
```

```
association_results[0]
```

```
for item in association_results:
  pair = item[0]
  items = [x for x in pair]
  print("Rule: " + items[0] + "->" + items[1])
  print("Support: " + str(item[1]))
  print("Confidence: " + str(item[2][0][2]))
  print("Lift: " + str(item[2][0][3]))
  print("\n")
```
